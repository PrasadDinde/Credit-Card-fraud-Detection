### Under Sampling

```python
normal = df[df['Class']==0]
fraud = df[df['Class']==1]

print(normal.shape)
print(fraud.shape)

normal_sample = normal.sample(n=473)
print(normal_sample.shape)

new_df = pd.concat([normal_sample,fraud],ignore_index=True)
print(new_df['Class'].value_counts())
new_df.head()
```

```python
x = new_df.drop('Class',axis=1)
y = new_df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
classifier = {
    "Logistic Regression " : LogisticRegression(),
    "Decision Tree Classifier ": DecisionTreeClassifier()
}

for name,clf in classifier.items():
    print(f"\n========{name}========")
    clf.fit(x_train,y_train)
    pred_val = clf.predict(x_test)
    accuracy = accuracy_score(y_test,pred_val)
    print(f"\n Accuracy: {accuracy_score(y_test,pred_val)}")
    print(f"\n Precision: {precision_score(y_test,pred_val)}")
    print(f"\n Recall: {recall_score(y_test,pred_val)}")
    print(f"\n F1_Score: {f1_score(y_test,pred_val)}")
    
```
